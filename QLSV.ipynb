```python
class Student:
    def __init__(self, ma_sv, ho_ten, ngay_sinh, que_quan):
        self.ma_sv = ma_sv
        self.ho_ten = ho_ten
        self.ngay_sinh = ngay_sinh
        self.que_quan = que_quan
        
    def info(self):
        print(self.ma_sv.ljust(20) + self.ho_ten.ljust(20) + self.ngay_sinh.ljust(20) + self.que_quan.ljust(20))
        print('-'*68)
        
class ListStudent:
    def __init__(self, danh_sach_sv):
        self.danh_sach_sv = danh_sach_sv
        
    def load(self):
        sv1 = Student('001', 'Lộc Nguyễn', '10/08/1987', 'TP.HCM')
        sv2 = Student('002', 'Thủy Phạm', '21/05/1987', 'Đà Lạt')
        sv3 = Student('009', 'Nam Trần', '25/09/1980', 'TP.HCM')
        self.danh_sach_sv = [sv1, sv2, sv3]
    
    def them(self, sv):
        self.danh_sach_sv.append(sv)
    
    def xoa(self, ma_sv):
        result = False
        for sv in self.danh_sach_sv:
            if sv.ma_sv == ma_sv: 
                self.danh_sach_sv.remove(sv)
                result = True
                break
        return result
                
                
    def sua(self, ma_sv, ho_ten, ngay_sinh, que_quan):
        result = False
        for sv in self.danh_sach_sv:
            if sv.ma_sv == ma_sv:
                sv.ho_ten = ho_ten
                sv.ngay_sinh = ngay_sinh
                sv.que_quan = que_quan
                result = True
                break
            return result
                
    def info(self):
        print('Mã sinh viên'.ljust(20) + 'Họ tên'.ljust(20) + 'Ngày sinh'.ljust(20) + 'Quê quán'.ljust(20))
        print('-'*68)
        for sv in self.danh_sach_sv:
            sv.info()
        print(' ')
                
class Program:
    def __init__(self, dssv=None):
        self.dssv = ListStudent(None)
        self.dssv.load()
        
    def run(self):
        while 1:
            print('='*10 + 'QUẢN LÝ SINH VIÊN' + '='*10)
            print('1. Hiển thị danh sách')
            print('2. Thêm sinh viên')
            print('3. Chỉnh sửa thông tin')
            print('4. Xóa thông tin')
            print('5. Tìm kiếm thông tin')
            print('6. Sắp xếp theo tên')
            print('7. Thoát')
            print('='*37)
            opt = input('Nhập lựa chọn: ')
            if opt == '1':
                self.dssv.info()
            elif opt == '2':
                ma_sv = input('Nhập mã sinh viên: ')
                ho_ten = input('Nhập họ tên: ')
                ngay_sinh = input('Nhập ngày sinh: ')
                que_quan = input('Nhập quê quán: ')
                sv = Student(ma_sv, ho_ten, ngay_sinh, que_quan)
                self.dssv.them(sv)
                print(f'Thêm thành công mã sinh viên {ma_sv}')
            elif opt == '4':
                ma_sv = input('Nhập mã sinh viên cần xóa: ')
                if self.dssv.xoa(ma_sv) == True:
                    print(f'Xóa thành công mã sinh viên {ma_sv}')
                else:
                    print(f'Không tìm thấy mã sinh viên {ma_sv}')
            elif opt == '7':
                confirm = input('Thoát chương trình? (y/n) ')
                if confirm == 'y' or confirm == 'Y':
                    break;

prog = Program()
prog.run()
    
# student_list = ListStudent(None)
# student_list.load()

# student_list.them(Student('003', 'Khải Nguyên', '04/10/1989', 'Lâm Đồng'))
# student_list.info()

# student_list.sua('003', 'Châu Huỳnh', '01/04/1994', 'Đồng Tháp')
# student_list.info()

# student_list.xoa('001')
# student_list.info()

# student_list.them(Student('007', 'Lộc Nguyễn', '10/08/1987', 'TP.HCM'))
# student_list.info()
```
